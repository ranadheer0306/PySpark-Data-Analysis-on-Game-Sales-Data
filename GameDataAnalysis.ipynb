#Which year has the most Games released?

```python
df.groupBy("Year").count().orderBy(F.col("count").desc()).show(5)
```

#Top 5 years with count of games in each genre

```python
yearRank = df.groupBy('Year').agg(F.count('*').alias('yearCount')) \
             .withColumn('YearRank', F.row_number().over(Window.orderBy(F.desc('yearCount')))).filter(F.col('YearRank')<=5).drop('yearCount')
```

```python
yearRank.show()
```

```python
getGenresCount = df.join(yearRank,on=['Year'],how='inner').orderBy(F.col('YearRank')).groupBy('Year').pivot('Genre').count()
yearRanks = yearRank.groupBy('Year').agg(F.first('YearRank').alias('Rank'))
```

```python
finaOp = getGenresCount.join(yearRanks,on=['Year'],how='inner').orderBy(F.col('Rank')).drop('Rank')
```

```python
finaOp.show()
```

#Which Year had the highest sales worldwide?

```python
df.groupBy('Year') \
.agg(F.round(F.sum('Global_Sales'),2) \
     .alias("Sum_of_sales")) \
     .orderBy(F.col("Sum_of_sales") \
              .desc()).show(5)
```

#Which genre game has been released the most in a single year?


```python
df.groupBy('Year','Genre').count() \
.orderBy(F.col("count") \
                 .desc()).show()
```

#Which genre game has sold the most in a single year?


```python
df.groupBy('Year', 'Genre')\
.agg(F.round(F.sum('global_sales'),3).alias('sum_of_sales')) \
.groupBy('Year') \
.agg(F.max('sum_of_sales').alias("Total_year_sales")) \
.orderBy(F.col('Total_year_sales').desc()).show()
```

#Which genre game have the highest sale price globally?

```python
df.groupBy('Genre').agg(F.round(F.sum("Global_sales"),3).alias("sum_of_sales")).orderBy(F.col("sum_of_sales").desc()).show()
```

#Which platfrom have the highest sale price globally?

```python
df.groupBy('Platform').agg(F.round(F.sum("Global_sales"),3).alias("sum_of_sales")).orderBy(F.col("sum_of_sales").desc()).show()
```

#Which individual game have the highest sale price globally?

```python
df.orderBy(F.col('Global_sales').desc()).select('Name', 'Global_Sales').first()
```

#Top Global Sales of the Game of each Publisher

```python
df.groupBy("Publisher").agg(F.max('Global_sales').alias('Total_sales')).orderBy(F.col('Total_sales').desc()).show(20)
```

```python
df.groupBy('Publisher', 'Name').agg(F.max('Global_sales').alias('Total_sales')).withColumn('TopPublishers',
                                                        F.row_number().over(Window.partitionBy('Publisher').orderBy(F.col('Total_sales').desc()))).\
                                                        filter(F.col('TopPublishers')==1).orderBy(F.col('Total_sales').desc()).drop('TopPublishers').show()
```
